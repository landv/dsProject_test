```python
import twitter

from secret import API_KEY, API_SECRET_KEY, ACCESS_TOKEN, ACCESS_TOKEN_SECRET
```

```python
api = twitter.Api(consumer_key=API_KEY,
                      consumer_secret=API_SECRET_KEY,
                      access_token_key=ACCESS_TOKEN,
                      access_token_secret=ACCESS_TOKEN_SECRET)
```

```python
results = api.GetSearch(term='#apple', count=100)
```

```python
for r in results:
    if r.retweet_count > 20:
        print(r.created_at)
        print(r.retweet_count)
        print(r.text)
        print('-'*150)
```
