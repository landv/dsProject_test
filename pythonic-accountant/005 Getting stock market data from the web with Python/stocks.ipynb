```python
import yfinance as yf
import altair as alt
import pandas as pd
```

```python
data = yf.download('MSFT', period='5y')
```

```python
len(data)
```

```python
data.tail()
```

```python
data.loc[pd.datetime(2018, 12, 31)].Close
```

```python
price = alt.Chart(data.reset_index()).mark_line().encode(
    x='Date:T',
    y='Close',
    tooltip=['Date', 'Open','Close', 'High', 'Low', 'Volume']
).properties(
    width=600, height=200)

volume = alt.Chart(data.reset_index()).mark_line().encode(
    x='Date:T',
    y='Volume'
).properties(
    width=600, height=60)

price & volume
```
