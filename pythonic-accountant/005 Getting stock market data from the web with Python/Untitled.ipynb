```python
import yfinance as yf
import altair as alt
import pandas as pd
```

```python
data = yf.download('AAPL', period='5y')
```

```python
# data.tail()
```

```python
data.loc[pd.datetime(2018, 12, 18)].Close
```

```python
price = alt.Chart(data.reset_index()).mark_line().encode(
    x='Date:T',
    y='Close',
    tooltip=['Date', 'Open', 'Close', 'High', 'Low', 'Volume'],
).properties(
    width=600, height=200
)

volume = alt.Chart(data.reset_index()).mark_line().encode(
    x='yearmonth(Date):T',
    y='mean(Volume)'
).properties(
    width=600, height=60
)

price & volume
```
