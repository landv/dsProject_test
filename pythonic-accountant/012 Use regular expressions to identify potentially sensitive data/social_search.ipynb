```python
import tarfile
import re
from datetime import datetime
```

```python
ssn = re.compile(r'\d{3}-\d{2}-\d{4}')
email = re.compile(r'[A-Za-z0-9]+@[A-Za-z0-9]+\.[A-Za-z0-9]+')
```

```python
tar = tarfile.open("enron_mail_20150507.tar.gz", "r")
```

```python
items = tar.getmembers()
```

```python
# '4566 you order your accounting firm to delete the emails and shred the'
# '41290 er your accounting firm to delete the emails  and shred the "evidence".  In='
```

```python
for idx, member in enumerate(items[:1000]):
    f = tar.extractfile(member)
    try:
        content = f.read().decode() if f else ''
    except:
        pass

    if not content:
        continue

    date = content.split('\n')[1]
    date_parsed = datetime.strptime(' '.join(date.split('Date: ')[1].split()[:-2]), '%a, %d %b %Y %H:%M:%S')
    
    if (date_parsed.year == 2001) and (date_parsed.month == 10) and (date_parsed.day == 12):
        
        print(content)

```
