```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('countries of the world.csv')
```

```python
data.rename(columns={'Country':'country','Region':'region','Area (sq. mi.)':'area','Population':'population','Area (sq. mi.)':'area','Pop. Density (per sq. mi.)':'population_density','Coastline (coast/area ratio)':'coastline','Net migration':'migration','Infant mortality (per 1000 births)':'infant_mortality','GDP ($ per capita)':'gdp','Literacy (%)':'literacy','Phones (per 1000)':'phones','Arable (%)':'arable','Crops (%)':'crops','Other (%)':'other','Climate':'climate','Birthrate':'birthrate','Deathrate':'deathrate','Agriculture':'agriculture','Industry':'industry','Service':'service'},inplace=True)
```

```python
data.columns
```

```python
#converting the datatypes into category and float for benefits
data.country=data.country.astype('category')
data.region=data.region.astype('category')
data.population_density=data.population_density.str.replace(",",".").astype(float)
data.coastline=data.coastline.str.replace(",",".").astype(float)
data.migration=data.migration.str.replace(",",".").astype(float)
data.infant_mortality=data.infant_mortality.str.replace(",",".").astype(float)
data.literacy=data.literacy.str.replace(",",".").astype(float)
data.phones=data.phones.str.replace(",",".").astype(float)
data.arable=data.arable.str.replace(",",".").astype(float)
data.crops=data.crops.str.replace(",",".").astype(float)
data.other=data.other.str.replace(",",".").astype(float)
data.climate=data.climate.str.replace(",",".").astype(float)
data.birthrate=data.birthrate.str.replace(",",".").astype(float)
data.deathrate=data.deathrate.str.replace(",",".").astype(float)
data.agriculture=data.agriculture.str.replace(",",".").astype(float)
data.industry=data.industry.str.replace(",",".").astype(float)
data.service=data.service.str.replace(",",".").astype(float)
```

```python
data.info()
```

```python
#removing the Nan values and replacing with them mean,removing extra spaces from the region column
data.fillna(data.mean(),inplace=True)
data.region=data.region.str.strip()
```

```python
data.head(10)
```

```python
#number of countries in a region
region=data.region.value_counts()
sns.barplot(x=region.index,y=region.values)
plt.ylabel('Number of countries')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.title('Number of countries by region',color = 'red')
```

```python
#distribution of birth rate
sns.distplot(data['birthrate'])
```

```python
y=data.corr()
sns.heatmap(y,cmap='autumn',linewidths=0.8)
```

```python
#most populated countries
most_populated_countries = data.sort_values('population',ascending=False).head(20)
mean = pd.DataFrame({'country':['data mean'], 'population':[data['population'].mean()]})
most= pd.concat([most_populated_countries[['country','population']],mean],ignore_index=True)
sns.barplot(x='country',y='population',data=most, palette='Set3')
plt.xticks(rotation=90)
```

```python
#most literacy rate
data.groupby('country')['literacy'].sum().sort_values(ascending=False).head()
```

```python
#top gdp countries
top_gdp_countries = data.sort_values('gdp',ascending=False).head(20)
mean = pd.DataFrame({'country':['data mean'], 'gdp':[data['gdp'].mean()]})
gdps = pd.concat([top_gdp_countries[['country','gdp']],mean],ignore_index=True)
sns.barplot(x='country',y='gdp',data=gdps, palette='Set3')
plt.xticks(rotation=90)
```

```python
#distribution of infant mortality rate
sns.distplot(data['infant_mortality'],color='red')
plt.title("Distribution of Infant Mortality Rate")
```

```python
#relation of gdp with other parameters
f, ax = plt.subplots(3,3, figsize = (18,10))
sns.scatterplot(data = data, x = "gdp", y = "literacy", ax = ax[0,0])
sns.scatterplot(data = data, x = "gdp", y = "birthrate", ax = ax[0,1])
sns.scatterplot(data = data, x = "gdp", y = "deathrate", ax = ax[0,2])
sns.scatterplot(data = data, x = "gdp", y = "infant_mortality", ax = ax[1,0])
sns.scatterplot(data = data, x = "gdp", y = "agriculture", ax = ax[1,1])
sns.scatterplot(data = data, x = "gdp", y = "industry", ax = ax[1,2])
sns.scatterplot(data = data, x = "gdp", y = "climate", ax = ax[2,0])
sns.scatterplot(data = data, x = "gdp", y = "service", ax = ax[2,1])
sns.scatterplot(data = data, x = "gdp", y = "phones", ax = ax[2,2])
plt.show()
```
